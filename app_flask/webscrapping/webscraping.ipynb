```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.esportsearnings.com/history/2025/games/164-league-of-legends"


headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'}
response = requests.get(url,headers=headers)

player_data = []

if response.status_code == 200:
    soup = BeautifulSoup(response.text,'html.parser')

    soup_allplayer_info = soup.find_all(class_="detail_list_table")
    print(soup_allplayer_info)

    for player in soup_allplayer_info:
        info_player = player.find_all(class_="format_row")
        print(info_player)
        for row in info_player:
            cols = row.find_all("td")
            data = [col.get_text(strip=True) for col in cols]

            country_tag = row.find("a") 
            country = "Unknown"

            if country_tag:
                if "title" in country_tag.attrs:
                    country = country_tag["title"]
                else:
                    baliseImg = country_tag.find("img")
                    if baliseImg and "title" in baliseImg.attrs: 
                        country = baliseImg["title"]
                    elif baliseImg and "alt" in baliseImg.attrs:
                        country = baliseImg["alt"]

            data.append(country)

            print(data)

            player_data.append(data)


print(player_data)
```

```python
player_data[1][0] = "Rang"
player_data[1][-1] = "Country"
print(player_data[1][0])
column_names = player_data[1]
data_rows = player_data[2:]
df = pd.DataFrame(data_rows, columns=column_names)

```

```python
df = df.dropna()
```
